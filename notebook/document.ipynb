```python
### let us create a basic   directory inside the data folder
import os
os.makedirs("../data/text_files", exist_ok=True) 
### the difference between mkdir and makedirs is that mkdir will crash if prior file path is not present but on the other hand the makedirs will create every other file that is not present in the file path.
```

```python
### sample files is bascially a dictionary where the key is the file path and the value is the content of the file. We are creating two text files with some content in them.
sample_files = {
    "../data/text_files/python_intro.txt" : """
    Python Programming Introduction

Python is a high-level, interpreted programming language known for its simplicity and readability.
Created by Guido van Rossum and first released in 1991, Python has become one of the most popular
programming languages in the world.

Key Features:
- Easy to learn and use
- Extensive standard library
- Cross-platform compatibility
- Strong community support

Python is widely used in web development, data science, artificial intelligence, and automation.
""",

"../data/text_files/machine_learning.txt" : """
Machine learning is a subset of artificial intelligence that enables systems to learn and improve
from experience without being explicitly programmed. It focuses on developing computer programs
that can access data and use it to learn for themselves.

Types of Machine Learning:
1. Supervised Learning: Learning with labeled data
2. Unsupervised Learning: Finding patterns in unlabeled data
3. Reinforcement Learning: Learning through rewards and penalties

Applications include image recognition, speech processing, and recommendation systems
"""
}

### now below is the for loop using the key,value pair of the sample files dictionary that is used to create the respective textfile with "w" i.e, write format.
for filepath,content in sample_files.items():
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(content)
```

```python
### here we can use the either to import the Loader (here basically we are converting various file formats which containt he text into a document structure) from the langchain_community or the langchain. Both are same but the langchain_community is more updated and has more features than the langchain. So it is better to use the langchain_community for loading the documents.
from langchain_community.document_loaders import TextLoader
### (from langchain.document_loaders import TextLoader) we cannot use thsi as this has been removed from the langchain and is now only available in the langchain_community.

loader = TextLoader("../data/text_files/python_intro.txt",encoding="utf-8")
document = loader.load()
print(document)
```

```python
#now let us look how to load a whole directory at once
from langchain_community.document_loaders import DirectoryLoader
dir_loader = DirectoryLoader(
    "../data/text_files", ## this tells the location of the dir in which you should find the files to be loaded.
    glob = "**/*.txt", ## (*.txt) this tell the compiler to grab all the text files in the dir with file type .txt and (**) this tells to look for the text files in the subdirectories as well if there are any.
    loader_cls = TextLoader, ## this tells the compiler to use the TextLoader i.e, specifically  to load the files. We can also use other loaders like PDFLoader, CSVLoader etc depending on the file type we are loading.
    loader_kwargs = {"encoding" : "utf-8"}, ## this is the keyword argument for the loader class i.e, the TextLoader in this case. Here we are specifying the encoding type of the text files to be loaded.
    show_progress = False ## this is to show the progress of the loading process. If there are many files to be loaded then it is better to set this to True to see the progress of the loading process.
)
documents = dir_loader.load()
print(documents)
```
